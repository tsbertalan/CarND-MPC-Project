```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class T:
    def __init__(self, theta, xt, yt, unit='angle'):
        if unit == 'angle':
            theta = theta * np.pi / 180.
        c = np.cos(theta)
        s = np.sin(theta)
        self.A = np.array([
            [c, -s, xt],
            [s,  c, yt],
            [0,  0, 1 ],
        ])
        self.theta = theta
        self.xt = xt
        self.yt = yt
        
    def __call__(self, xa, ya):
        X = np.array([[xa, ya, 1]]).T
        return tuple(self.A.dot(X)[:2].ravel())
    
    def show(self):
        fig, ax = plt.subplots(figsize=(5,5))
        
#         X = np.linspace(0, 1, 42)
#         Y = np.linspace(0, 2, 42)
#         X, Y = np.meshgrid(X, Y)
#         X = X.ravel()
#         Y = Y.ravel()
        
        X = np.array([0, 1, 1, 0, 0])
        Y = np.array([0, 0, 2, 2, 0])
        
        ax.plot(X, Y, label='original')
        
        Xb, Yb = self(X, Y)
        ax.plot(Xb, Yb, label='transformed')
        
        ax.set_title(r'$\theta=%s\pi$, $x_t=%s$, $y_t=%s$' % (self.theta/np.pi, self.xt, self.yt))
        
        xl = list(ax.get_xlim())
        yl = list(ax.get_ylim())
        
        l = [] + xl + yl
        
        ax.set_xlim(min(l), max(l))
        ax.set_ylim(min(l), max(l))
        
        ax.legend()
        

T(45, 1, 2).show()
T(4, 1, 2).show()
T(4, 3, 2).show()
```
